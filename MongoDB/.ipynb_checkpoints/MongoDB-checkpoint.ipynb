```python
import pymongo

ip = 'localhost'
port = 27017

# Connect MongoDB
connection = pymongo.MongoClient(ip, port)
# ['Test_Database', 'admin', 'config', 'local']
print(connection.list_database_names())

# Connect MongoDB Database
database = connection.get_database('test')
# ['Test_Collection']
print(database.list_collection_names())

# Connect MongoDB Database Collection
collection = database.get_collection('monsters')

# find Data
result = collection.find()
for i in result:
    print(i)

```
